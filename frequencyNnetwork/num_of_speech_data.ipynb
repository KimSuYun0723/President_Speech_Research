# 대통령별 연설문 개수 시각화

# 1.필요한 라이브러리 다운로드 & 설치, 폰트 설정

```python
# 시각화 라이브러리: seaborn, matplotlib
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 윈도우 시스템에 있는 Malgun Gothic 폰트를 사용
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

# 2. 대통령별 연설문 수

```python
# 대통령 이름 리스트
name_file = ['LeeSeungMan', 'YoonBoSun', 'ParkJeongHee', 'ChoiGyuHa', 'JeonDuHwuan',
             'RohTaeWoo', 'KimYoungSam', 'KimDaeJung', 'RohMooHyun', 'LeeMyungBak',
             'ParkGeunHye', 'MoonJaeIn', 'YoonSeokYeol']
kor_name = {'LeeSeungMan':'Rhee Syng-man', 'YoonBoSun':'Yun Po-sun', 'ParkJeongHee':'Park Chung-hee',
            'ChoiGyuHa':'Choi Kyu-hah', 'JeonDuHwuan':'Chun Doo-hwan', 'RohTaeWoo':'Roh Tae-woo',
            'KimYoungSam':'Kim Young-sam', 'KimDaeJung':'Kim Dae-jung', 'RohMooHyun':'Roh Moo-hyun',
            'LeeMyungBak':'Lee Myung-bak', 'ParkGeunHye':'Park Geun-hye', 'MoonJaeIn':'Moon Jae-in', 'YoonSeokYeol':'Yoon Suk-yeol'}

# 데이터 저장을 위한 리스트
president_data = []

# 데이터 개수 계산
for name in name_file:
    file_path = f'./dataset/cleaned_data/cleaned_{name}.csv'
    president_df = pd.read_csv(file_path, encoding='utf-8')
    data_count = len(president_df)
    president_data.append({'대통령': kor_name[name], '데이터 수': data_count})

# 데이터프레임 생성
df_counts = pd.DataFrame(president_data)

# 데이터프레임 출력
print(df_counts)

# 시각화
plt.figure(figsize=(12, 6))
bars = plt.bar(df_counts['대통령'], df_counts['데이터 수'], color='skyblue')
plt.xticks(rotation=45, ha='center')
plt.xlabel('Presidents', labelpad=15)
plt.ylabel('Number of Speech Data')
plt.title('Number of Speech Data by President')
plt.tight_layout()

# 막대 위에 데이터 수 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, int(yval), ha='center', va='bottom')

plt.savefig('presidents_speech_count.jpg', dpi=300, bbox_inches='tight')  # 이미지 저장
plt.show()
```
